##### State wise Minimum Average Pricing of the Cars

```python
cars.head()
```

```python
cars['state'].unique()
```

```python
#A Need to pull all prices of statewise data
#B Sum it to Sum of Pricing State
#C Number of cars per state
#D Divide Sum by Number of cars for Average Pricing
```

```python
states = pd.DataFrame({
    'state':cars['state'].unique()
})

# Get Sum for one state
newJersy = cars[cars['state'] == 'new jersey']
newJersy.head()
# states.head()
```

```python
newJersy['price'].sum()
```

```python
states['sum'] , states['noOfCars'] = 0,0
# For loop to get sum of all states with number of Cars in each state
for itr in states.index:
    localState = cars[cars['state'] == states['state'][itr]]
    noOfCars = len(localState)
    localSum = localState['price'].sum()
    #print(f"{states['state'][itr]} -> {localSum}")
    states['sum'][itr] , states['noOfCars'][itr] = localSum,noOfCars
    
states.head()
```

```python
# let find the average pricing per state 
states['avgPricing'] = states['sum'] / states['noOfCars'];
states.head()
```

```python
states.sort_values(by="avgPricing").head()
```
